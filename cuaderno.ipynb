```python
import pandas as pd
```

```python
dataframe = pd.read_csv('./df_anscombe.csv')
```

```python
dataframe.groupby('group').describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
grupos = dataframe['group'].unique()
num_grupos = len(grupos)

# Crear subplots dinámicamente según el número de grupos
fig, axes = plt.subplots(1, num_grupos, figsize=(5 * num_grupos, 5), sharex=True, sharey=True)

# Convertir a array si hay un solo grupo
if num_grupos == 1:
    axes = [axes]

# Iterar sobre los grupos y graficar los puntos
for i, grupo in enumerate(grupos):
    subset = dataframe[dataframe['group'] == grupo]
    ax = axes[i]
    
    ax.scatter(subset['x'], subset['y'], color='blue', edgecolor='black', s=80)
    
    ax.set_title(f'Grupo {grupo}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear un colormap de Seaborn para colores únicos por grupo
palette = sns.color_palette("tab10", len(grupos))

# Crear la figura
plt.figure(figsize=(8, 6))

# Iterar sobre los grupos y graficar cada uno con un color diferente
for i, grupo in enumerate(grupos):
    ax = axes[i]
    subset = dataframe[dataframe['group'] == grupo]
    plt.scatter(subset['x'], subset['y'], color=palette[i], label=f'Grupo {grupo}', edgecolor='black', s=80)
    sns.regplot(x='x', y='y', data=subset, ax=ax, scatter_kws={'color': 'blue', 's': 120}, line_kws={'color': 'red'})

# Configurar la gráfica
plt.title("Grupos Diferenciados por Color")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(title="Grupo")
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear una paleta de colores única por grupo
palette = sns.color_palette("tab10", len(grupos))

# Crear la figura
plt.figure(figsize=(8, 6))

# Iterar sobre los grupos y graficar cada uno con su color
for i, grupo in enumerate(grupos):
    subset = dataframe[dataframe['group'] == grupo]
    sns.regplot(
        x='x', y='y', data=subset, 
        scatter_kws={'color': palette[i], 's': 60, 'edgecolor': 'black'}, 
        line_kws={'color': palette[i]}, 
        label=f'Grupo {grupo}'
    )

# Configurar la gráfica
plt.title("Regresión por Grupo en una Sola Gráfica")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(title="Grupo")
plt.grid(True)

# Mostrar la gráfica
plt.show()
```

```python
###Linea de regresion
fig, axes = plt.subplots(1, num_grupos, figsize=(5 * num_grupos, 5), sharex=True, sharey=True)

if num_grupos == 1:
    axes = [axes]

# Iterar sobre los grupos y graficar la regresión
for i, grupo in enumerate(grupos):
    subset = dataframe[dataframe['group'] == grupo]
    ax = axes[i]
    
    # Gráfico de dispersión con línea de regresión
    sns.regplot(x='x', y='y', data=subset, ax=ax, scatter_kws={'color': 'blue', 's': 120}, line_kws={'color': 'red'})
    
    # Ajustar títulos y etiquetas
    ax.set_title(f'Regresión Grupo {grupo}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True)

plt.tight_layout()
plt.show()
```
